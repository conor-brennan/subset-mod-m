```python
import numpy as np
import random
```

```python
def soln(x, Soln):
    if x<=0:
        return []
    else:
        return Soln[x]

def subset_sum_mod_m(S,m):
    n = len(S)
    Soln = [[]]*n
    if n<=0:
        return f"n > 0 required, given {n}"
    
    for i in range(n):
        Soln[i] = soln(i-1,Soln).copy()
        if S[i]%m not in Soln[i]:
            Soln[i].append(S[i]%m)
        for v in soln(i-1,Soln):
            if (v+S[i])%m not in Soln[i]:
                Soln[i].append((v+S[i])%m)
    return Soln

def reconstruct_zero_mod_m(S,m,Soln):
    index = np.inf
    for i in range(len(Soln)):
        if 0 in Soln[i]:
            index = i
            break
    if index == np.inf:
        return f"Soln does not contain 0 (mod {m})"

    R = [S[i]]
    goal = -S[i]%m
    while goal!=0:
        i=i-1
        if goal in Soln[i] and goal not in soln(i-1, Soln):
            goal = (goal - S[i])%m
            R.append(S[i])
    return R

def arr_sum(arr):
    sum = 0
    for arri in arr:
        sum+=arri
    return sum
```

```python
for i in range(500):
    # Pick random modulo
    modulo = random.randint(1, 20)
    # Pick a random length array of numbers in [0..m-1]
    S = [random.randint(0,modulo-1) for _ in range(random.randint(1, 20))]
    # Compute Soln array
    soln_arr = subset_sum_mod_m(S,modulo)
    # Find 0 reconstruction (mod m)
    R = reconstruct_zero_mod_m(S,modulo,soln_arr)
    if isinstance(R, str):
        print(f"For m={modulo}, S={S}, {R}")
    else:
        print(f"For m={modulo}, S={S}, the sum of {R}={arr_sum(R)}=0(mod {modulo})")
```
